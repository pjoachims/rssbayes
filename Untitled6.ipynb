```python
# Rosenbrock's banana function
def banana(z):
    z = z.T
    a = 1.
    b = 100.
    return (a-z[0])**2 + b*(z[1] - z[0]**2)**2

def banana_log_p(z):
    return -banana(z)

def himmelblau(z):
    z = z.T
    return (z[0]**2 + z[1] - 11)**2 + (z[0] + z[1]**2 - 7)**2

def himmelblau_log_p(z):
    return -himmelblau(z)

def eggholder(z):
    z = z.T
    x = z[0]
    y = z[1]
    return -(y + 47)*np.sin((np.abs(x/2 + y + 47))**0.5) - x*np.sin((np.abs(x + y - 47))**0.5)

def eggholder_log_p(z):
    return -eggholder(z)

z = tt.matrix('z')
z.tag.test_value = pm.floatX([[0., 0.]])
bananaf = theano.function([z], banana(z))
himmelblauf = theano.function([z], himmelblau(z))
```

```python
sample_size = 1000
SEED = 12345

alpha = 0.1

with pm.Model() as pot1m:
    pm.DensityDist('pot1', logp=banana_log_p, shape=(2,))
    trace1 = pm.sample(sample_size, step=pm.Metropolis(), random_seed=SEED)
    trace2 = pm.sample(sample_size, step=pm.Slice(), random_seed=SEED)
    trace3 = pm.sample(sample_size, step=pm.NUTS(), random_seed=SEED)
    
_, ax = plt.subplots(1,3,figsize=(10,4),sharex=True, sharey=True)

tr1 = trace1['pot1']
ax[0].plot(tr1[:,0], tr1[:,1], 'bo--',alpha=alpha)
ax[0].title.set_text('Metropolis')
   
tr2 = trace2['pot1']
ax[1].plot(tr2[:,0], tr2[:,1], 'go--',alpha=alpha)
ax[1].title.set_text('Gibbs (almost)')

tr3 = trace3['pot1']
ax[2].plot(tr3[:,0], tr3[:,1], 'ro--',alpha=alpha)
ax[2].title.set_text('NUTS')

plt.tight_layout()
plt.savefig("plot_banana.pdf")
```
