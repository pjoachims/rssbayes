```python
from matplotlib import pyplot as plt
import numpy as np
import pymc3 as pm
import seaborn as sns

%matplotlib inline
```

```python
np.random.seed(42) # from random.org, for reproducibility
```

```python
N = 1000

W = np.array([0.35, 0.4, 0.25])

MU = np.array([0., 2., 5.])
SIGMA = np.array([0.5, 0.5, 1.])
```

```python
component = np.random.choice(MU.size, size=N, p=W) #randomly selecting an index for MU/SIGMA
x = np.random.normal(MU[component], SIGMA[component], size=N) #generating the data
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.hist(x, bins=30, normed=True, lw=0);
```

```python
with pm.Model() as model:
    w = pm.Dirichlet('w', np.ones_like(W))
    mu = pm.Normal('mu', 0., 10., shape=W.size)
    tau = pm.Gamma('tau', 1., 1., shape=W.size)
    x_obs = pm.NormalMixture('x_obs', w, mu, tau=tau, observed=x)
```

```python
with model:
    trace = pm.sample(5000, n_init=10000, tune=1000, random_seed=42)
```

```python
pm.traceplot(trace, var_names=['w', 'mu']);
```

```python
with model:
    ppc_trace = pm.sample_posterior_predictive(trace, 5000, random_seed=42)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

ax.hist(ppc_trace['x_obs'], bins=30, density=True,
        histtype='step', lw=2,
        color=['.5'] * ppc_trace['x_obs'].shape[1],
        alpha=.05,
        label='Posterior predictive distribution');

ax.hist(x, bins=30, density=True,
        histtype='step', lw=2,
        label='Observed data');

ax.legend(loc=1);
```
